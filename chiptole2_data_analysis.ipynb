```python
chipo=pd.read_csv('chiptole2.csv')
chipo.head(5)
```

### Step 4. How many products cost more than $10.00?

```python
# clean the item_price column and transform it in a float
# reassign the column with the cleaned prices
chipo['item_price']=chipo['item_price'].apply(lambda x: x.replace('$','')).astype(float)
chipo['item_price']
```

```python
# delete the duplicates in item_name and quantity and choice_description
chipo2=chipo.drop_duplicates(subset=["item_name","quantity","choice_description"])
chipo2
```

```python
# select only the products with quantity equals to 1

chipo2=chipo2[chipo2['quantity']==1]
chipo2
```

```python
# products cost more than $10.00
df=chipo2[chipo2['item_price']>10]
df['item_price'].unique()
```
